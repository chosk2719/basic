## pie chart

```python
# ex -12-9

pets = ['개', '고양이', '기타', '기르지않음']
portion=[18,3.4,3.1,75.5]

plt.pie(portion, explode = (0, 0.1, 0, 0), labels = pets, autopct = '%1.1f%%', shadow = True, startangle=90)
plt.title('총 가구 대비반려동물 기르는 비율(2018)')

plt.show()
```

```python
# Q12-1
data = list(range(1, 11))
mim_temp = [7.3, 3.5, 7.2, 4.5, 3.5, 8.3, 7.7, 6.4, 9.3, 6.3]
max_temp = [15, 16.3, 15.2, 17.5, 18.3, 20.0, 18.8, 17.3, 14.2, 15.3]

plt.plot(data, mim_temp, label = 'Min Temparature', color = 'red', linestyle = '-', marker='o')
plt.plot(data, max_temp, label = 'Max Temparature', color = 'blue', linestyle = '-.', marker='x')
plt.legend(loc = 'best')
plt.xlabel('data')
plt.ylabel('temparature')
plt.title('Min and max temparature of ten days')

plt.show()
```

```python
# Q12-2

xdata = ['안지영', '홍지수', '황예림']
ydata1 = [90, 85, 88]
ydata2 = [83, 88, 91]
ydata3 = [85, 97, 78]
ydata4 = [92, 88, 82]

plt.plot(xdata, ydata1,label = '국어', color = 'red', linestyle = '--', marker='*')
plt.plot(xdata, ydata2,label = '영어', color = 'green', linestyle = '-', marker='s')
plt.plot(xdata, ydata3,label = '수학', color = 'blue', linestyle = ':', marker='d')
plt.plot(xdata, ydata4,label = '사회', color = 'cyan', linestyle = '-.', marker='o')

plt.ylim(0,100)
plt.title('세명 학생의 네 과목 성적')
plt.legend(loc = 'best')

plt.show()
```

## stem plot

```python
x = np.linspace(0.1, 2 * np.pi, 10)
plt.title("Stem Plot")
plt.stem(x, np.cos(x), '-.')
plt.show()
```

```python
N = 30
np.random.seed(0)
x = np.random.rand(N)
y1 = np.random.rand(N)
y2 = np.random.rand(N)
y3 = np.pi * (15 * np.random.rand(N))**2
plt.title("Bubble Chart")
plt.scatter(x, y1, c=y2, s=y3)
plt.show()
```

## Image(imshow)

```python
from sklearn.datasets import load_digits
digits = load_digits()
X = digits.images[0]
X
```

```python
plt.title("mnist digits; 0")
plt.imshow(X, interpolation='nearest', cmap=plt.cm.bone_r)
plt.xticks([])
plt.yticks([])
plt.grid(False)
plt.subplots_adjust(left=0.35, right=0.65, bottom=0.35, top=0.65)
plt.show()
```

```python
#등고선

def f(x, y):
    return (1 - x / 2 + x ** 5 + y ** 3) * np.exp(-x ** 2 - y ** 2)


n = 256
x = np.linspace(-3, 3, n)
y = np.linspace(-3, 3, n)
XX, YY = np.meshgrid(x, y)
ZZ = f(XX, YY)

plt.title("Contour plots")
plt.contourf(XX, YY, ZZ, alpha=.75, cmap='jet')
plt.contour(XX, YY, ZZ, colors='black')
plt.show()
```

```python
def f(x, y):
    return (1 - x / 2 + x ** 5 + y ** 3 ) * np.exp(-x ** 2 - y **2)


n = 256
x = np.linspace(-3,3,n)
y = np.linspace(-3,3,n)
XX, YY = np.meshgrid(x,y)
ZZ= f(XX, YY)

plt.title('Contour plot')
plt.contourf(XX,YY,ZZ,alpha=.75, cmap='jet')
plt.contourf(XX,YY,ZZ,color='black')
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
X = np.arange(-4, 4, 0.25)
Y = np.arange(-4, 4, 0.25)
XX, YY = np.meshgrid(X, Y)
RR = np.sqrt(XX**2 + YY**2)
ZZ = np.sin(RR)

fig = plt.figure()
ax = Axes3D(fig)
ax.set_title("3D Surface Plot")
ax.plot_surface(XX, YY, ZZ, rstride=1, cstride=1, cmap='hot')
plt.show()
```
