```python
# ex 12-4

xdata = ['안지영', '홍지수', '황예린']
ydata1 = [90, 85, 88]
ydata2 = [83, 88, 91]
plt.plot(xdata, ydata1, label='국어')
plt.plot(xdata, ydata2, label='영어')
plt.legend(loc='upper center')

plt.title('세명 학생의 국어, 영어 성적')

plt.show()
```

```python
# 12-5 Line style

xdata = ['안지영', '홍지수', '황예린']
ydata1 = [90, 85, 88]
ydata2 = [83, 88, 91]
ydata3 = [85, 97, 78]
ydata4 = [92, 88, 82]

plt.plot(xdata, ydata1, label='국어',color='red', linestyle='-', marker='o')
plt.plot(xdata, ydata2, label='영어',color='#00ffff', linestyle='--', marker='x')
plt.plot(xdata, ydata3, label='수학',color='magenta', linestyle='-.', marker='s')
plt.plot(xdata, ydata4, label='사회',color='#444444', linestyle=':', marker='d')


plt.title('세명 학생의 네 과목 성적')
plt.legend(loc='best')

plt.show()
```

```python
plt.plot(xdata, ydata1, 'ro-', label='국어', linewidth=6, markersize=10)
plt.plot(xdata, ydata2, 'cx--',label='영어',markersize=10)
plt.plot(xdata, ydata3, 'ms-.',label='수학',markersize=10)
plt.plot(xdata, ydata4, 'kd:',label='사회',markersize=10)

plt.title('세명 학생의 네 과목 성적')
plt.legend()
plt.show()
```

```python
# 12-6

xdata = list(range(30))
ydata = []

for x in xdata:
    y = 2*x
    ydata.append(y)
    
plt.title('y=2x')
plt.plot(xdata,ydata, color='magenta', linewidth=3) #label='y=2x')
plt.xlim(-1,30)
plt.ylim(-2,60)
plt.xticks(list(range(0,30,2)))
plt.yticks(list(range(0,60,5)))

plt.grid(True)
plt.show()
```

```python
x = numpy.arange(0,20,0.1)
y1 = numpy.sin(x)
y2 = numpy.cos(x)

plt.plot(x, y1, label='sin')
plt.plot(x, y2, linestyle='--', label='cos')
plt.xlabel('x')
plt.xlabel('y')
plt.title('sin &amp; cos')
# plt.xticks(list(range(0,30,1)))
# plt.yticks(list(range(0,60,1)))
plt.legend()
plt.show()
```

```python
xs = numpy.linspace(-3,3,601)
ys = 1 / (1 + numpy.exp(-xs))
plt.plot(xs, ys)
plt.tlib(-0.1, 1.1)
plt.title('시그모이드 그래프')
plt.axhline(y=0,color='k', linewidth=0.5)
plt.axhline(y=1,color='k', linewidth=0.5)
plt.axvline(x=0,color='k', linewidth=0.5)
plt.show()
```

```python
def f(x, y):
    return (1 - x / 2 + x ** 5 + y ** 3) * np.exp(-x ** 2 - y ** 2)


n = 256
x = np.linspace(-3, 3, n)
y = np.linspace(-3, 3, n)
XX, YY = np.meshgrid(x, y)
ZZ = f(XX, YY)

plt.title("Contour plots")
plt.contourf(XX, YY, ZZ, alpha=.75, cmap='jet')
plt.contour(XX, YY, ZZ, colors='black')
plt.show()
```
