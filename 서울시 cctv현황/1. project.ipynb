## 외국인 비율과 고령자 비율을 추가

```python
pop['외국인 비율'] = pop['외국인'] / pop['인구수'] * 100
pop['고령자 비율'] = pop['고령자'] / pop['인구수'] * 100
pop.head()
```

## 6. 최근 증가율 데이터

```python
# cctv = cctv.fillna(0), astype(int)
```

```python
cctv[['2011년 이전', '2012년', '2013년']] = \
    cctv[['2011년 이전', '2012년', '2013년']].fillna(0).astype(int)
# cctv['2011년 이전'] = cctv['2011년 이전'].fillna(0).astype(int)
# cctv['2012년 이전'] = cctv['2012년'].fillna(0).astype(int)
# cctv['2013년 이전'] = cctv['2013년'].fillna(0).astype(int)
cctv
```

```python
cctv['2015년 이전 합계'] = cctv[['2011년 이전', '2012년', '2013년', '2014년', '2015년']].sum(axis=1)
cctv
```

```python
# cctv['zxcv'] = cctv.iloc[:,2:7].sum(axis=1)
# cctv.head()
```

```python
cctv['최근 3년 개수'] = cctv[['2016년', '2017년', '2018년']].sum(axis=1)
cctv['최근 증가율'] = cctv['최근 3년 개수'] / cctv['2015년 이전 합계'] * 100
cctv
```

### 7.자치구, 소계, 최근 증가율 칼럽외 모두 삭제

```python
cctv.drop(['2011년 이전', '2012년', '2013년', '2014년', '2015년'], axis = 1, inplace = True)
cctv.head()
```

```python
del cctv ['2016년']
del cctv ['2017년']
del cctv ['2018년']
del cctv ['최근 3년 개수']
del cctv ['2015년 이전 합계']
cctv.head()
```

## 8. 서울인구 데이터, cctv데이터 합치고, '자치구'를 인덱스로 만든다.

```python
df = pd.merge(cctv, pop, on='자치구')
df.head()
```

```python
df.set_index('자치구', inplace=True)
df.head()
```

## 9.cctv현황을 다양한형태로 시각화

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rcParams['axes.unicode_minus'] = False
mpl.rc('font', family='Malgun Gothic')
%matplotlib inline
```

```python
plt.rcParams["figure.figsize"] = (10,6)
plt.rcParams["font.size"] = 14
```

### 자치구별 cctv 댓수 (내림차순으로)

```python
df['소계'].sort_values().plot(kind='barh', grid=True, figsize=(12,12))
plt.show()
```

```python
df['인당 cctv'] = df['소계'] / df['인구수']
df['인당 cctv'].sort_values().plot(kind='barh', grid=True, figsize=(12,12))
plt.show()
```

#### - 자치구별 최근 증가율(내림차순으로)

```python
df['최근 증가율'].sort_values().plot(kind='barh', grid=True, figsize=(12,12))
plt.show()
```

#### - 인구수와 CCTV댓수의 산점도

```python
plt.figure(figsize=(12,10))
plt.scatter(df['인구수'], df['소계'], s=50)
plt.xlabel('인구수')
plt.ylabel('CCTV대수')
plt.grid()
plt.show()
```

#### - 인구수와 CCTV댓수의 산점도 및 회귀선

```python
fp1 = np.polyfit(df['인구수'], df['소계'], 1)
fp1
```

```python
f1 = np.poly1d(fp1)
fx = np.linspace(100000, 700000, 101)
```

```python
plt.figure(figsize=(12,10))
plt.scatter(df['인구수'], df['소계'], s=50)
plt.plot(fx, f1(fx), ls='dashed', lw=3,color='m')
plt.xlabel('인구수')
plt.ylabel('CCTV대수')
plt.grid()
plt.show()
```

#### - 인구수와 CCTV댓수의 산점도, 회귀선 및 자치구명

```python
df.head()
```

```python
df['오차'] = np.abs(df['소계'] - f1(df['인구수']))
df.head()
```

```python
df_sort = df.sort_values(by='오차', ascending=False)
df_sort.head(10)
```

```python
plt.figure(figsize=(14,10))
plt.scatter(df['인구수'], df['소계'], c= df['오차'], s=50)
plt.plot(fx, f1(fx), ls='dashed', lw=3,color='g')

for n in range(10):
    plt.text(df_sort['인구수'][n]*1.02,df_sort['소계'][n]*0.98,
            df_sort.index[n], fontsize=15)



plt.xlabel('인구수')
plt.ylabel('cctv 대수')
plt.colorbar()
plt.grid()
plt.show()
```
